```python
entries_df = pd.read_csv(os.path.join(BASE_FILE_PATH, 'entries.csv'))
entries_df.head()
```

```python
merged_df = df.merge(entries_df, on='summoner_id', how='inner')

account_distribution = merged_df.groupby(['tier', 'rank']).agg({'summoner_id': 'nunique'}).reset_index()

account_distribution.columns = ['Tier', 'Rank', 'Number of Accounts']

account_distribution = account_distribution.sort_values(by=['Tier', 'Rank'])

print(account_distribution)
```

```python
def assign_time_group(days):
    if days <= 7:
        return 'Active Players (1-7 days)'
    elif 8 <= days <= 14:
        return 'Slightly Inactive (8-14 days)'
    elif 15 <= days <= 28:
        return 'Moderately Inactive (15-28 days)'
    elif 29 <= days <= 40:
        return 'Highly Inactive (29-40 days)'
    elif 41 <= days <= 60:
        return 'At Risk of Churn (41-60 days)'
    else:
        return 'Beyond 60 days - Lost'


merged_df['game_creation_dt'] = pd.to_datetime(merged_df['game_creation_dt'])

# Find the maximum date in the 'game_creation_dt' column
max_date = merged_df['game_creation_dt'].max()

latest_game_per_summoner = merged_df.groupby('summoner_id')['game_creation_dt'].max().reset_index()
latest_game_per_summoner.rename(columns={'game_creation_dt': 'latest_game_dt'}, inplace=True)

latest_game_per_summoner['days_since_last_game'] = (max_date - latest_game_per_summoner['latest_game_dt']).dt.days

merged_df = merged_df.merge(latest_game_per_summoner[['summoner_id', 'days_since_last_game']], on='summoner_id', how='left')

merged_df['time_group'] = merged_df['days_since_last_game'].apply(assign_time_group)

merged_df[['summoner_id', 'game_creation_dt', 'days_since_last_game', 'time_group']].head()



```

```python
# Group by 'tier', 'rank', and 'time_group', counting the number of unique summoner_ids in each group
grouped_data = merged_df.groupby(['tier', 'rank', 'time_group'])['summoner_id'].nunique().reset_index()

# Rename columns for clarity
grouped_data.columns = ['Tier', 'Rank', 'Time Group', 'Number of Accounts']

# Pivot the table to get a more readable format, if desired
pivot_table = grouped_data.pivot_table(index=['Tier', 'Rank'], columns='Time Group', values='Number of Accounts', fill_value=0)

# Display the resulting table
pivot_table
```

```python
def assign_binary_time_group(days):
    if days <= 28:
        return 'Active'
    else:
        return 'Inactive'

latest_game_per_summoner['binary_time_group'] = latest_game_per_summoner['days_since_last_game'].apply(assign_binary_time_group)

latest_game_per_summoner

```

```python
# Group by 'tier', 'rank', and 'time_group', counting the number of unique summoner_ids in each group
grouped_data = merged_df.groupby(['tier', 'rank', 'binary_time_group'])['summoner_id'].nunique().reset_index()

# Rename columns for clarity
grouped_data.columns = ['Tier', 'Rank', 'binary_time_group', 'Number of Accounts']

# Pivot the table to get a more readable format, if desired
pivot_table = grouped_data.pivot_table(index=['Tier', 'Rank'], columns='binary_time_group', values='Number of Accounts', fill_value=0)

# Display the resulting table
pivot_table
```
